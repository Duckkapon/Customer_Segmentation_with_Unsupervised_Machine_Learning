# K-Mean Clustering
Unsupervised หมายถึงเป็น Model ที่ใส่ Input Data เข้าไปโดยที่เราไม่ได้กำหนด Target ของโมเดลโดยจะให้คอมพิวเตอร์เป็นคนกำหนดเองว่าข้อมูลที่ได้รับมาสามารถแปลงออกมาเป็นโมเดลในลักษณะรูปแบบใดได้บ้าง 
 
โดยกระบวนการทำ Clustering เป็นการนำเอาข้อมูลมาจัดกลุ่ม โดยข้อมูลที่อยู่ในกลุ่มเดียวกันจะมีความสัมพันธ์หรือมีลักษณะที่คล้ายคลึงกัน เช่น การจัดกลุ่มลูกค้าที่มีลักษณะความต้องการที่เหมือนกัน หรือจัดกลุ่มสินค้าที่มีคุณลักษณะเหมือนกัน ซึ่งสอดคล้องการวัตถุประสงค์ของเราในการทำ project นี้
 
โดยความสำคัญของการทำ Clustering Model เนื่องจากขั้นตอนที่สำคัญสำหรับองค์กรคือ การทำตลาดซึ่งถือเป็นหนึ่งในหัวใจของการขายสินค้า ซึ่งการที่จะทำการตลาดกับลูกค้านั้นเราไม่สามารถตอบสนองกลุ่มเป้าหมายทุกคนได้ ดังนั้นเพื่อให้การทำการตลาดเข้าถึงกลุ่มเป้าหมายจำเป็นต้องมีการทำ Segmentation เพื่อจัดกลุ่มลูกค้าที่มีลักษณะคล้ายคลึงกันเพื่อให้เกิดการรับรู้ถึงสินค้าและบริการ และจะทำให้เราสามารถรู้หน้าตาของกลุ่มลูกค้าแต่ละกลุ่มได้

\
การทำ clustering มีหลักง่าย ๆ คือ ข้อมูลที่มีความคล้ายกัน ถูกจัดอยู่ใน cluster เดียวกัน

### หลักการในการแบ่งกลุ่ม Clustering
การจัดข้อมูลให้อยู่ในกลุ่มต่างๆ จะต้องมีวัดค่าคล้ายคลึง (similarity) หรือ ค่าระยะห่าง (Euclidean distance) ระหว่างข้อมูลแต่ละตัว
1. กำหนดหรือสุ่มค่าเริ่มต้น จำนวน K ค่า (กลุ่ม)
2. กำหนดจุดศูนย์กลางเริ่มต้น K จุด เรียกว่า cluster centers หรือ Centroid
3. นำวัตถุทั้งหมดจัดเข้ากลุ่ม โดยทำการหาค่าระยะห่างระหว่างข้อมูล 
4. กับจุดศูนย์กลาง หากข้อมูลไหนใกล้ค่าจุดศูนย์กลางตัวไหนที่สุดอยู่ กลุ่มนั้น
5. หาค่าเฉลี่ย แต่ละกลุ่ม ให้เป็นค่าจุดศูนย์กลางใหม่ ( new centroid )
6. ทำซ้ำข้อ 2 จนกระทั่งค่าเฉลี่ยหรือจุดศูนย์กลางในแต่ละกลุ่มจะไม่เปลี่ยนแปลงจึงหยุดทำ

## Optimize dataframe memory usage

```python
def reduce_mem_useage(df) :
    print("start".center(50,"="))
    print(f"Before optimize memory usage = {df.memory_usage().sum() / 1024**2} MB")
    for col in df.columns :
        org = df[col]
        trans = df[col].astype(np.uint8)
        check = (org == trans).min()
        if check == True :
            df[col] = trans
        else :
            trans = df[col].astype(np.uint16)
            check = (org == trans).min()
            if check == True :
                df[col] = trans
            else :
                trans = df[col].astype(np.uint32)
                check = (org == trans).min()
                if check == True :
                    df[col] = trans
                else :
                    trans = df[col].astype(np.int8)
                    check = (org == trans).min()
                    if check == True :
                        df[col] = trans
                    else :
                        trans = df[col].astype(np.int16)
                        check = (org == trans).min()
                        if check == True :
                            df[col] = trans
                        else :
                            trans = df[col].astype(np.int32)
                            check = (org == trans).min()
                            if check == True :
                                df[col] = trans
                            else :
                                df[col] = df[col].astype(np.float32)
    print(f"After optimize memory usage = {df.memory_usage().sum() / 1024**2} MB")
    print("done".center(50,"="),"\n")
```

need to extract summary data for each customers

```python
df_trans = output.copy()
df_trans.info()
```

```python
pd.set_option('display.float_format', '{:20,.2f}'.format)
```

```python
# Pivot sum and mean, amount and quantity
count_transaction = df_trans.groupby(["customer_id"])["transaction"].count().astype("int16")
pivot = pd.DataFrame()
for col in ["amount", "quantity"] :
    if col == "amount" :
        name = "Amt"
    else :
        name = "Qty"
    for agg in ["sum", "mean"] :
        pv = pd.pivot_table(df_trans, values=col, index='customer_id', columns='shop_id', aggfunc=agg, fill_value = 0)
        if agg == "sum" :
            pv.columns = [f"shop{col}Total{name}" for col in pv.columns]
            pv[f"allShopTotal{name}"] = np.sum(pv, axis = 1)
        else :
            pv.columns = [f"shop{col}Mean{name}" for col in pv.columns]
        pivot = pd.concat([pivot, pv], axis = 1)
pivot = pd.merge(pivot, count_transaction, on = "customer_id", how="inner")
pivot.reset_index(inplace = True)
```

```python
pivot.info()
```

```python
# find AOV 
for i in range(1,6) :
    cols = [f"shop{i}TotalAmt", "transaction"]
    df = pivot[cols].copy()
    df[f"shop{i}AOV"] = df[cols[0]]/df[cols[1]]
    df[f"shop{i}AOV"].fillna(0, inplace=True)
    # display(df[[f"shop{i}AOV"]])
    pivot = pd.concat([pivot, df[[f"shop{i}AOV"]]], axis = 1)

pivot["allShopAOV"] = pivot["allShopTotalAmt"]/pivot["transaction"]
pivot["allShopAOV"].fillna(0, inplace=True)
```

## Auto find K

Source: https://jtemporal.com/kmeans-and-elbow-method/

```python
# Within cluster sum square error
# calcualte error from each cluster

# range of K
min_range = 2
max_range = 16

def calculate_wcss(data):
    wcss = []
    for n in range(min_range, max_range):
        kmeans = KMeans(n_clusters=n,random_state=0)
        kmeans.fit(model)
        wcss.append(kmeans.inertia_)
    
    return wcss

from math import sqrt
#calculate 
def optimal_number_of_clusters(wcss):
    x1, y1 = min_range, wcss[0]
    x2, y2 = max_range, wcss[len(wcss)-1]

    distances = []
    for i in range(len(wcss)):
        x0 = i+2
        y0 = wcss[i]
        numerator = abs((y2-y1)*x0 - (x2-x1)*y0 + x2*y1 - y2*x1)
        denominator = sqrt((y2 - y1)**2 + (x2 - x1)**2)
        distances.append(numerator/denominator)
    
    return distances.index(max(distances)) + 2
    

# calculating the within clusters sum-of-squares for n cluster amounts
sum_of_squares = calculate_wcss(model)
    
# calculating the optimal number of clusters
n = optimal_number_of_clusters(sum_of_squares)
print('Number fo cluster =', n)
```

## Fit model

k = 6

```python
KMeans = KMeans(n_clusters=n, random_state=0)
KMeans.fit(model)

final_df["cluster"] = KMeans.predict(model)
```
